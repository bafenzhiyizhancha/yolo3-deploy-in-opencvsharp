#### 图像拷贝整理

获取到的数据这样：

目的：将拍照数据拷贝到新文件夹JPEGImages下：

使用c#编写如下代码，其中路径需要修改成你,同时由于需要排列的图片名字格式是"0152",四位数,所以需要从至少100开始计算,否则代码规则会变复杂难以控制.

仅为示例,可根据自己需求编写,唯一目的是将获取的图片保存在同一文件夹下,其中图片的命名没有特殊要求.

```python
private void GetPicture()
{
    string path = "D:/Woring/yolo3/50N1/50N1/2021-10-12";
    string HDMIName = "HDMI2.jpg";
    string savePath = "D:/Woring/yolo3/JPEGImages";
    int num = 100;
    string imageName = "";
    string saveName = "";

    string[] folders = Directory.GetDirectories(path);

    foreach(var name in folders)
    {
        imageName = name + "/"+ HDMIName;
        Bitmap image = new Bitmap(imageName);

        saveName = savePath + "/0" + num.ToString() + ".jpg";
        image.Save(saveName);

        num = num + 1;
    }
}
```

此处以上和以下是c#代码，仅以提供示例为目的，也可使用python编写脚本、手工人力操作，达到目的即可。

## 训练

使用本次配置好的代码即可，点击yolo3下的train.py

对原代码修改的地方如下所示，共4处，同时也是需要注意的点：


1、394行，epoch：是训练迭代次数，建议一般设为270次。<br/>
2、395行，batch_size:是一次迭代输入的图像数量，在此由于本计算机内存小，故设为8，如果是GPU训练，可以设为16或者32。<br/>
3、409行，device，训练是cpu还是gpu。<br/>

重要参数说明参数：

1、394行，epoch：是训练迭代次数，建议一般设为270次。<br/>
2、395行，batch_size:是一次迭代输入的图像数量，在此由于本计算机内存小，故设为8，如果是GPU训练，可以设为16或者32。<br/>
3、396行，cfg:是配置参数文件的路径，已将配置好了，请勿修改。<br/>
4、397行，data：是训练数据的输入地址,将配置好了，请勿修改。<br/>
5、399行，img-size，实验过，该参数请勿修改。<br/>
6、409行，device，训练是cpu还是gpu。<br/>

在初始阶段建议将[394]行的epchs的default设为10，训练完成，检测一下是否满足要求，满足则将该值设为200，继续训练
